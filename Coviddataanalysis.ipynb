## lets create TreeMap Representation of our data

```python
df.columns
```

```python
grouped_data = df.groupby('Country/Region')[['Confirmed', 'Deaths', 'Recovered', 'Active']].sum()
```

```python
for col in grouped_data.columns:
    fig = px.treemap(grouped_data.sort_values(by=col, ascending=False).reset_index(),
                     values=col, path=['Country/Region'], template="plotly_white",
                     title=f"<b>TreeMap representation of Countries w.r.t. {col}</b>")
    fig.show()
```

What is the % distribution of cases among various WHO regions?

```python
# Group by 'WHO Region' and sum 'Confirmed', 'Deaths', 'Recovered', and 'Active'
Who_regions = df.groupby('WHO Region')[['Confirmed', 'Deaths', 'Recovered', 'Active']].sum().sort_values('Confirmed', ascending=False)

# Display the result
print(Who_regions)
```

```python
# Group by 'WHO Region' and sum 'Confirmed', 'Deaths', 'Recovered', and 'Active'
Who_regions = df.groupby('WHO Region')[['Confirmed', 'Deaths', 'Recovered', 'Active']].sum()

# Select the top 15 regions based on the 'Confirmed' cases
top_15_who_regions = Who_regions.sort_values('Confirmed', ascending=False).head(15)

# Plot pie charts for each case type (Confirmed, Deaths, Recovered, Active)
cases = ['Confirmed', 'Deaths', 'Recovered', 'Active']
for case in cases:
    fig = px.pie(top_15_who_regions, values=case, names=top_15_who_regions.index,
                 template="plotly_dark", hole=0.3, 
                 title=f"{case} Recorded w.r.t. WHO Region of 15 Worst Affected Regions")
    fig.show()
```

```python
grouped_data_months = df.groupby('month_name').agg({'Confirmed': 'sum',
                                                            'Deaths': 'sum',
                                                            'Recovered': 'sum',
                                                            'Active': 'sum',
                                                            'New cases': 'sum',
                                                            'New deaths': 'sum',
                                                            'New recovered': 'sum'})

```

```python
grouped_data_months
```

```python

# Define the correct month order as a mapping of month names to their respective numbers
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
month_num_mapping = {month: i+1 for i, month in enumerate(month_order)}

# Add a new column for sorting based on the month number
grouped_data_months_reset['month_num'] = grouped_data_months_reset['month_name'].map(month_num_mapping)

# Sort the data based on the 'month_num'
grouped_data_months_reset = grouped_data_months_reset.sort_values('month_num')

# Create line chart with Plotly
fig = px.line(grouped_data_months_reset, x='month_name', y=grouped_data_months_reset.columns[1:], 
              title='COVID-19 Cases by Month', labels={'month_name': 'Month'},
              template='plotly_white')

fig.update_layout(
    xaxis_title='Month',
    yaxis_title='Count',
    xaxis_tickangle=-45,  # Rotate x-axis labels for better readability
    legend_title='Metrics'
)

fig.show()
```

```python
#group data by country
country_grouped = df.groupby('Country/Region').max()
country_grouped.sample(5)
```

```python
import plotly.express as px

# Top 10 countries with the highest confirmed cases
top10_confirmed = country_grouped.nlargest(10, 'Confirmed')
fig_confirmed = px.bar(
    top10_confirmed,
    x='Confirmed',
    y=top10_confirmed.index,
    orientation='h',
    color='Confirmed',
    title='Top 10 Countries with the Highest Confirmed Cases',
    labels={'Confirmed': 'Number of Confirmed Cases', 'y': 'Country'},
    color_continuous_scale='BrBG'
)
fig_confirmed.update_layout(
    xaxis_title='Number of Confirmed Cases (in millions)',
    yaxis_title='Country',
    template='plotly_dark'
)
fig_confirmed.show()

# Top 10 countries with the highest number of deaths
top10_deaths = country_grouped.nlargest(10, 'Deaths')
fig_deaths = px.bar(
    top10_deaths,
    x='Deaths',
    y=top10_deaths.index,
    orientation='h',
    color='Deaths',
    title='Top 10 Countries with the Highest Deaths',
    labels={'Deaths': 'Number of Deaths', 'y': 'Country'},
    color_continuous_scale='RdBu'
)
fig_deaths.update_layout(
    xaxis_title='Number of Deaths',
    yaxis_title='Country',
    template='plotly_dark'
)
fig_deaths.show()



```
